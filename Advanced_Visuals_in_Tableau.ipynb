Question 1 : Create a Heatmap to analyze Treatment Cost by Department and Region.

- Answer 1

To identify where the most money is being spent, you need to aggregate the costs by the specific medical condition.

Logic: This helps hospital administrators understand which diseases are the most "expensive" to treat at scale.

Tableau Steps:

1.  Drag Diagnosis to the Rows shelf.

2.  Drag Treatment Cost to the Columns shelf (it will default to SUM).

3.  Sort the bars in Descending order.

4.  Hold Ctrl and drag Diagnosis to the Filters shelf. Select the Top tab, choose By field, and set it to Top 5 by Treatment Cost (Sum).

Question 2 : Create a Highlight Table showing Satisfaction Score by Department.

- Answer 2

This metric evaluates patient experience across different wings of the healthcare facility.

Logic: A low score in a high-cost department might indicate a need for process improvement.

Tableau Steps:

    1. Drag Department to Rows.

    2.  Drag Satisfaction Score to Columns.

    3.  Right-click the Satisfaction Score pill in the shelf and change the measure from Sum to Average.

    4.  Tip: Use a Color Mark to highlight departments falling below a certain threshold.

Question 3 : Create a Bump Chart to show ranking of Departments based on Treatment Cost over months.

- Answer 3

Visualizing where your patients come from helps in resource allocation and identifying regional health trends.

Logic: If the "North" region has a spike in "Asthma," the facility might need more respiratory specialists in that area.

Tableau Steps:

1.  Double-click Region.

2.  Drag PatientID to the Size or Color mark.

3.  Change the measure of PatientID to Count (Distinct) to show the volume of patients per region.

Question 4 : Create a Butterfly Chart comparing Male vs Female patient count by Department.

- Answer 4

This analyzes whether treatment becomes significantly more expensive as patients get older.

Logic: Look for "clusters." If costs are high for young patients, it may be due to chronic conditions like Cancer; if high for elderly, it may be age-related surgeries.

Tableau Steps:

1.  Drag Age to Columns.

2.  Drag Treatment Cost to Rows.

3.  Go to the Analysis menu and uncheck Aggregate Measures to see individual patient dots.

4.  Optional: Drag Diagnosis to Color to see if specific diseases drive the cost-age relationship.
